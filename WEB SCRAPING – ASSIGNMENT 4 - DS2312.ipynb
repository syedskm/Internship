1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: 

A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
#setting a driver
driver = webdriver.Chrome()
```

```python
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos#Top_videos')
```

```python
vid = []
for i in driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]//td[1]'):
    vid.append(i.text)
```

```python
ytr = []
for i in driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]//td[2]'):
    ytr.append(i.text)
```

```python
upl = []
for i in driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]//td[4]'):
    upl.append(i.text)
```

```python
view = []
for i in driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]//td[3]'):
    view.append(i.text)
```

```python
ran = []
for i in range(1,len(upl)+1):
    ran.append((i))
```

```python
import re
video = []
for i in vid:
    name = re.sub(r'\[\d+\]', '', i)
    video.append(name)
```

```python
youtube = pd.DataFrame({'Rank':ran,'Name':video,'Artist':ytr,'Upload Date':upl,'Views':view})
```

have to find the following details:

C) Task 
D) Attribute type 
E) No of instances 
F) No of attribute G) Year 

```python
#setting a driver
driver = webdriver.Chrome()
```

```python
driver.get('https://archive.ics.uci.edu/')
```

```python
all_sets = driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div/div/div/a[1]')
all_sets.click()
```

```python
ds = []
dt = []
tsk = []
inst = []

for i in driver.find_elements(By.XPATH,'//h2[@class="truncate text-primary"]/a'):
    ds.append(i.text)
for i in driver.find_elements(By.XPATH,'//div[@class="col-span-3 flex items-center gap-2"][2]'):
    dt.append(i.text)
for i in driver.find_elements(By.XPATH,'//div[@class="col-span-3 flex items-center gap-2"][1]'):
    tsk.append(i.text)
for i in driver.find_elements(By.XPATH,'//div[@class="col-span-3 flex items-center gap-2"][3]'):
    inst.append(i.text)
```

```python
from selenium.webdriver.support import expected_conditions as EC
```

```python
dropdown_xpath = '//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]'


dropdown_elements = driver.find_elements(By.XPATH, dropdown_xpath)

for dropdown in dropdown_elements:
    try:
        driver.execute_script("arguments[0].scrollIntoView(true);", dropdown)
        
        WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, dropdown_xpath))) 

        dropdown.click()
    except Exception as e:
        print("An error occurred")
```

```python
yr = []
att = []
natt = []
for i in driver.find_elements(By.XPATH,'//tbody[@class="border"]//td[3]'):
    yr.append(i.text)
for i in driver.find_elements(By.XPATH,'//tbody[@class="border"]//td[2]'):
    att.append(i.text)
for i in driver.find_elements(By.XPATH,'//div[@class="col-span-3 flex items-center gap-2"][4]'):
    natt.append(i.text)
```

```python
datasets = pd.DataFrame({'Dataset name':ds,'Type':dt,'Year published':yr,'Task':tsk,'No of instances':inst,'Attribute type':att,'No of attribute':natt})
```
