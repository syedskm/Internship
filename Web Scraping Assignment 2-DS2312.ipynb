Q1. Write a python program to scrape data for ‚ÄúData Analyst‚Äù Job position in ‚ÄúBangalore‚Äù location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.
This task will be done in following steps:
1. First get the webpage https://www.shine.com/
2. Enter ‚ÄúData Analyst‚Äù in ‚ÄúJob title, Skills‚Äù field and enter ‚ÄúBangalore‚Äù in ‚Äúenter the location‚Äù field.
3. Then click the search button.
4. Then scrape the data for the first 10 jobs results you get.
5. Finally create a dataframe of the scraped data.

```python
#setting a driver
driver = webdriver.Chrome()
```

```python
#initialising website
driver.get('https://www.shine.com/')
```

```python
#to get the detailed search bar, click on the jobs search bar

des_dum = driver.find_element(By.XPATH,'/html/body/div/header/div[3]/div/div/div[1]/div/input')
des_dum.click()
```

```python
#searching Data Analyst roles
designation =  driver.find_element(By.CLASS_NAME,'form-control  ')
designation.click()
designation.send_keys('Data Analyst')

#searching for Bangalore location
location =  driver.find_element(By.XPATH,'/html/body/div/div[4]/div/div[2]/div[2]/div/form/div/div[1]/ul/li[2]/div/input')
location.send_keys('Bangalore')

#clicking on search
search = driver.find_element(By.XPATH,'/html/body/div/div[4]/div/div[2]/div[2]/div/form/div/div[2]/div/button')
search.click()
```

```python
#creating empty lists for job titles, location, experiance and company

job_title = []
job_location = []
experiance = []
company = []
```

```python
#scraping job title from shine.com
for i in driver.find_elements(By.XPATH,'//h2[@itemprop="name"]'):
    job_title.append(i.text)

#scraping job location from shine.com
for i in driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"]'):
    job_location.append(i.text)      

#removing \n
location=[]
for j in job_location:
    z = j.replace('\n', ' ')
    location.append(z)

#scraping job experiance from shine.com
for i in driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"]'):
    experiance.append(i.text)

#scraping company from shine.com
for i in driver.find_elements(By.XPATH,'//div[@class="jobCard_jobCard_cName__mYnow"]'):
    company.append(i.text)
```

```python
#creating dataframe for the scraped data
shine_data_analyst = pd.DataFrame({'Title':job_title[0:10], 'Company': company[0:10], 'Experiance': experiance[0:10], 'Location': location[0:10]})
```

```python
#initialising website
driver.get('https://www.amazon.in/')
```

```python
#searching for Laptop
search_bar = driver.find_element(By.XPATH,'/html/body/div/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input')
search_bar.send_keys('Laptop')

search = driver.find_element(By.XPATH,'/html/body/div/header/div/div[1]/div[2]/div/form/div[3]/div/span/input')
search.click()
```

```python
#creating empty lists for title, ratings, and price
title = []
ratings = []
price = []

#scraping data for title, ratings, and price
for i in driver.find_elements(By.XPATH,'//div[@data-cy="title-recipe"]'):
    title.append(i.text)
for i in driver.find_elements(By.CSS_SELECTOR,'span.a-icon-alt'):
    ratings.append(i.get_attribute('innerHTML'))
for i in driver.find_elements(By.CLASS_NAME,'a-price-whole'):
    price.append(i.text)
```

```python
#scraping stars alone, removing 'out of'
stars = []
for i in ratings:
    j = i.split('out of')
    stars.append(j[0])
```

```python
#taking top 10 data from each column

price = price[0:10]
stars = stars[2:12]
title = title[0:10]
```

```python
#creating dataframe for the scraped data
top_10_laptops = pd.DataFrame({'Title':title, 'Price': price, 'Ratings': stars})
```

Q8: Write a python program to scrape data for Top 1000 Quotes of All Time.
The above task will be done in following steps:
1. First get the webpagehttps://www.azquotes.com/
2. Click on Top Quotes
3. Than scrap a) Quote b) Author c) Type Of Quotes

```python
#setting a driver
driver = webdriver.Chrome()
```

```python
#initialising website
driver.get('https://www.azquotes.com/')
```

```python
#searching & clicking top 1000 quotes 
search = driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a')
search.click()
```

```python
#creating empty lists for quote, type, and author
quote = []
type = []
author = []

#scraping data for quote, type, and author
start = 0
end = 10

for page in range(start,end):
    for i in driver.find_elements(By.XPATH,'//a[@class="title"]'):
        quote.append(i.text)
    for i in driver.find_elements(By.CLASS_NAME,"tags"):
        type.append(i.text)
    for i in driver.find_elements(By.CLASS_NAME,'author'):
        author.append(i.text)
```

```python
#creating dataframe for the scraped data
quotes = pd.DataFrame({'Quote':quote, 'Author': author, 'Type of Quote': type})
```
