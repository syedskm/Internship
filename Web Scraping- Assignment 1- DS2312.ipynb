2. Write s python program to display list of respected former presidents of India(i.e. Name , Term ofoffice) from https://presidentofindia.nic.in/former-presidents.htm and make data frame.

```python
html = requests.get('https://presidentofindia.nic.in/former-presidents')
html
```

```python
soup_1 = BeautifulSoup(html.content)
```

```python
columns = soup_1.find_all('div', class_="desc-sec") 
```

```python
columns = soup_1.find_all('div', class_="desc-sec") 
pres = []
for i in columns:
    pres.append(i.text)
```

```python
president = []
for j in pres:
    name = j.split('\n')[1]
    time = j.split('\n')[2]
    president.append((name,time))
```

```python
presidents = pd.DataFrame(president, columns=["Name", "Tenure"])
presidents
```

3. Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape and make data framea) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

c) Top 10 ODI bowlers along with the records of their team andrating

a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
driver = webdriver.Chrome()
```

```python
driver.get('https://www.icc-cricket.com/rankings/team-rankings/mens/odi')
```

```python
odi_m = []
for i in driver.find_elements(By.CLASS_NAME,"si-table-row"):
    odi_m.append(i.text)
odi_men = []
```

```python
for j in odi_m:
    rank = j.split('\n')[0]
    team = j.split('\n')[1]
    match = j.split('\n')[2]
    point = j.split('\n')[3]
    rate = j.split('\n')[4]
    odi_men.append((rank,team,match,point,rate))
odi_men = odi_men[1:11]
```

```python
icc_men = pd.DataFrame(odi_men, columns=['Rank','Team','Matches','Points','Rating'])
icc_men
```

5. Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and make data frame-
i) Headline
ii) Time
iii) News Link

```python
url_news = requests.get('https://www.cnbc.com/world/?region=world')
url_news
```

```python
soup_news = BeautifulSoup(url_news.content)
```

```python
headli = []
for i in soup_news.find_all('a','LatestNews-headline'):
    headli.append(i.text)
```

```python
time = []
for i in soup_news.find_all('time','LatestNews-timestamp'):
    time.append(i.text)
```

```python
new_url = []
for i in soup_news.find_all('a','LatestNews-headline'):
    new_url.append(i['href'])
```

```python
cnbc_news = pd.DataFrame({'Headline':headli,'Time':time,'News Link':new_url})
cnbc_news
```

```python
header = soup.find_all('h2', class_="sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg")
```

```python
title = []
for i in header:
    title.append(i.text)
```

```python
wri = soup.find_all('span',class_="sc-1w3fpd7-0 dnCnAO")

author = []
for j in wri:
    author.append(j.text)
```

```python
dat = soup.find_all('span',class_="sc-1thf9ly-2 dvggWt")

released = []
for k in dat:
    released.append(k.text)
```

```python
lin = soup.find_all('a', class_="sc-5smygv-0 fIXTHm")

link = []
for l in lin:
    link.append(l['href'])
```

```python
Papers = pd.DataFrame({'Paper Title': title, 'Authors': author, 'Published Date': released, 'Paper URL': link})
```

7. Write a python program to scrape mentioned details from dineout.co.inand make data frame
i) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL

```python
do_url = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
```

```python
soup_4 = BeautifulSoup(do_url.content)

name = []

for i in soup_4.find_all('div', class_='restnt-info cursor'):
    name.append(i.text)
```

```python
cus = []

for i in soup_4.find_all('span', class_="double-line-ellipsis"):   
    cus.append(i.text)


cusin = []

for j in cus:
    price = j.split('|')[0]
    time = j.split('|')[1]
    cusin.append(time)
```

```python
loc = []

for i in soup_4.find_all('div', class_="restnt-loc ellipsis"):
    loc.append(i.text)
```

```python
rati = []

for i in soup_4.find_all('div', class_="img-wrap"):
    rati.append(i.text)
```

```python
img = []

for i in soup_4.find_all('img', class_="no-img"):
    img.append(i['data-src'])
```

```python
dineout = pd.DataFrame({'Restaurant':name,'Cusine':cusin,'Location':loc,'Ratings':rati,'Image URL':img})
dineout
```
