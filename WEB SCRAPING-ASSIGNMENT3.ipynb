2. In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then scrape all the products available under that product name. Details to be scraped are: "Brand
Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and
“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
#scraping product urls for the first 3 pages

prod_url = []
start = 0
end = 3
for page in range(start,end):
    driver.implicitly_wait(10)
    next_button = driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
    next_button.click()
    time.sleep(2)
    for i in driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]'):
        prod_url.append(i.get_attribute('href'))
```

```python
len(prod_url)
```

```python
url = prod_url[0:50]
```

```python
#scraping product name
product = []

for i in url:
    driver.get(i)

    for i in driver.find_elements(By.XPATH,'//span[@id="productTitle"]'):
        product.append(i.text)
```

```python
product
```

```python
#scraping price
price = []
for i in url:
    driver.get(i)
    for i in driver.find_elements(By.XPATH, '//span[contains(@class, "price")]'):
        price.append(i.text)
```

```python
price
```

```python
#removing \n
price_new = [i.replace('\n','.')for i in price]
price_new
```

```python
from selenium.common.exceptions import NoSuchElementException
```

```python
#scraping availability and return policy
avail = []
ret = []
for i in url:
    driver.get(i)
    for i in driver.find_elements(By.XPATH,'//div[@id="availability"]'):
        avail.append(i.text)
    for i in driver.find_elements(By.XPATH,'//*[@id="RETURNS_POLICY"]/span/div[2]/span'):
        ret.append(i.text)
```

```python
avail
```

```python
ret
```

```python
#scraping delivery date
del_date = []

for i in url:
    driver.get(i)
    for i in driver.find_elements(By.XPATH,'//*[@id="mir-layout-DELIVERY_BLOCK-slot-PRIMARY_DELIVERY_MESSAGE_LARGE"]/span/span[1]'):
        del_date.append(i.text)
```

```python
#scraping brand name
brand = []
for i in url:
    driver.get(i)

    try:
        brand_element = driver.find_element(By.XPATH, '//*[@id="productOverview_feature_div"]/div/table/tbody/tr[1]/td[2]/span')
        brand.append(brand_element.text)
    except NoSuchElementException:
        brand.append('--')
```

4. Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”,
“Secondary Camera”, “Display Size”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe and CSV.

```python
#setting a driver
driver = webdriver.Chrome()
```

```python
#initialising a website

driver.get('https://www.flipkart.com/')
```

```python
#searching a smart phone - 'Oneplus Nord'
search_bar =  driver.find_element(By.XPATH,'/html/body/div[1]/div/div[1]/div/div/div/div/div[1]/div/div[1]/div/div[1]/div[1]/header/div[1]/div[2]/form/div/div/input')
search_bar.click()
search_bar.send_keys('Oneplus Nord')

#clicking on search bar
search = driver.find_element(By.XPATH,'//*[@id="container"]/div/div[1]/div/div/div/div/div[1]/div/div[1]/div/div[1]/div[1]/header/div[1]/div[2]/form/div/button')
search.click()
```

```python
#scraping url
url = []
for i in driver.find_elements(By.XPATH,'//a[@class="_1fQZEK"]'):
    url.append(i.get_attribute('href'))
```

```python
#scraping price
price = []
for i in driver.find_elements(By.XPATH,'//div[@class="_30jeq3 _1_WHN1"]'):
    price.append(i.text)
```

```python
#scraping ram & rom
ram_rom = []
for i in driver.find_elements(By.XPATH,'//ul[@class="_1xgFaf"]/li[1]'):
    ram_rom.append(i.text)
```

```python
#creating individual lists for ram and rom

ram = []
rom = []
for i in ram_rom:
    ram.append(i.split('|')[0])
    rom.append(i.split('|')[1])
```

```python
#scraping name of the product
name = []
for i in driver.find_elements(By.XPATH, '//div[@class="_4rR01T"]'):
    name.append(i.text)
```

```python
#scraping brand name
brand = []
for i in name:
    k = i.split()
    brand.append(k[0])
```

```python
#scraping display size
disp = []
for i in driver.find_elements(By.XPATH,'//ul[@class="_1xgFaf"]/li[2]'):
    disp.append(i.text)
```

```python
#scraping battery capacity
battery = []
for i in driver.find_elements(By.XPATH,'//ul[@class="_1xgFaf"]/li[4]'):
    battery.append(i.text)
```

```python
#scraping camera details
camera = []
for i in driver.find_elements(By.XPATH,'//ul[@class="_1xgFaf"]/li[3]'):
    camera.append(i.text)
```

```python
#creating individual lists for primary cam and secondary cam
primary_camera = []
secondary_camera = []
for i in camera:
    if '|' in i:
        primary, secondary = i.split('|')
        primary_camera.append(primary.strip())
        secondary_camera.append(secondary.strip())
    else:
        primary_camera.append(i)
        secondary_camera.append('-')
```

```python
#scraping color
color = []
for j in url:
    driver.get(j)
    time.sleep(2)
    for i in driver.find_elements(By.XPATH,'//tr[@class="_1s_Smc row"][4]'):
        color.append(i.text)
```

```python
(color)
```

```python
#removing empty elements & splitting
colour = list(filter(None,color))
colour = [i.split('\n') for i in colour] 
```

```python
#storing the color alone and removing extra words
p_colour = [i[1] for i in colour]
```

```python
#creating a dataframe for the scraped data
flipkart_phone = pd.DataFrame({'Brand':brand,'Smartphone name':name,'Color':p_colour,'RAM':ram,'ROM':rom,'Primary Camera':primary_camera,'Secondary camera':secondary_camera,'Display':disp,'Battery capacity':battery,'Price':price,'Product URL':url})
```

```python
flipkart_phone
```

```python
#exporting the dataframe to csv
flipkart_phone.to_csv('flipkart_phone')
```

```python
#scraping no.of ratings and the overall rating keyword

no_rat = []
key_rat = []

for i in url:
    driver.get(i)
    for i in driver.find_elements(By.XPATH,'/html/body/div[3]/div/div/div[2]/section/div[8]/div/div[1]/div[1]/div[2]/div[1]/span'):
        no_rat.append(i.text)
    for i in driver.find_elements(By.XPATH,'//div[@class="keyword"]/span[1]'):
        key_rat.append(i.text)
```

```python
no_ratings = [no_rat[i] for i in range(len(no_rat)) if i % 2 == 0]      #there were every no.of ratings repeat, so removing the immediate repeat
no_ratings   #no of reviews
```

```python
keyword_ratings = [key_rat[i] for i in range(len(key_rat)) if i % 2 == 0]  #there were every keyword repeat, so removing the immediate repeat
keyword_ratings     #overall ratings 
```
